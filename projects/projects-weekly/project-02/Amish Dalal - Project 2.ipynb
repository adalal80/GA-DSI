## Cleaning the Genre Field

#### There are 24 artists that are associated with more than 1 unique genres. Thus we need to resolve this. For instance, Christina Aguilera would be considered Pop genre, but is tied to Rock and Rock n' Roll. To get proper genres, need to go through each artist and find out what their main genre would be.

```python
bb_genre = bb.groupby('artist')['genre'].nunique().to_frame()
print bb_genre[bb_genre['genre'] > 1].count()
bb_genre_fix = bb_genre[bb_genre['genre'] > 1]
bb_genre_fix
```

```python
artist = []
for x in bb_genre_fix.index:
    artist.append(x)
bb_genre_artist_fix = bb[bb['artist'].isin(artist)]
bb_genre_artist_fix
```
