# Project 2 - Evaluating Billboard Data Set

### Problem Statement

The billboard dataset that is provided is the Billboard Top 100 songs in the year 2000 by week. The study will evaluate Pop genre during Summer in the top 10 positions of Billboard Top 100. 


### Hypothesis 
#### Is there correlation with Pop hits and summer? Will Pop hits have more tracks in the billboard top 100?

Will pop have a higher rank in terms of position than the average mean in the summer?

### Hypothesis - 

H<sub>0</sub>: Average Position of Pop Genre during the Summer is equal to the mean of all genres during the summer

H<sub>a</sub>: Pop Genre during the Summer is not equal to the mean of all genres during the summer

### Assumptions

Assumption 1: one genre per artist. The data set genre fields must be cleaned up, as there are two variations of rock (rock and rock n roll) and two variations of R&B (R&B and R & B). The genre column was cleaned up as genre_mod in the csv itself, by going to Google and searching the genre of the artist.

Assumption 2: the time is the track length, which will be converted in total seconds

### Risks

1) There might be crossover between Pop and other Genres, such as Destiny's Child, which falls in R&B category, but some songs can fall into 'Pop' Genre

2) There are about a little less than 5% of tracks that dropped off the charts and came back on. Thus, this analysis will not count the inactive weeks in this analysis. 

3) The dataset shows only the tracks that peaked in 2000, thus has missing tracks when transformed into a dataset with weeks down the rows.

### Analyze
1) How do average Positions of tracks differ by Season and by Genre.

2) Look at climb/weeks to . is it higher during summer than other seasons? There is climb position and weeks to peak. look at the ratio for pop genre. Is there a difference between summer and rest of the seasons? Can use count of top 10 category to measure how many weeks in the top 10 category.

3) Look at dropped/(last week - week at peak) to analyze how fast it dropped. Pop genre should drop more slowly in the summer than other seasons



The dataset must be cleaned to ensure proper fields are in integer data types to conduct statistical analysis. Also Genre field must be cleaned up to show proper genre for a given artist. This will be done in the csv itself.

#### Cleaning the Billboard Dataset
1) Renaming artist column

2) Convert 'date_entered' and 'date_peaked' into datetime objects, as well as converting 'year' as a string

3) Converting 'time' into a seconds and storing it as a new column called 'track_length' in terms of seconds

4) Replace all '*' with np.nan

```python
#Renaming the artist column, changing genre to genre_old, and genre_mod
bb = bb.rename(columns = {'artist.inverted':'artist','genre':'genre_old','genre_mod':'genre'})
bb.columns = bb.columns.str.replace('.','_')

#converting date_entered and date_peaked into datetime objects
bb['date_entered'] = pd.to_datetime(bb['date_entered'])
bb['date_peaked'] = pd.to_datetime(bb['date_peaked'])

#Converting time column to a track length column in terms of seconds and integery type
#split time by separating the commas
l = []
for x in bb['time']:
    l.append(x.split(','))

#creating a new list of seconds by indexing on the 0th and 1st element by a for loop
sec = []
for x in l:
    minutes = int(x[0])
    seconds = int(x[1])
    total = minutes * 60 + seconds
    sec.append(total)
print sec
total_sec = pd.Series(sec)

#now that we have a list, lets find the location of the 'time' column 
#and insert a new column of total seconds to the right of it

bb.insert(bb.columns.get_loc('time')+1,'track_sec',total_sec)

#Writing a function to replace all '*' with np.nan 
def replace_nulls(value):
    if value == '*':
        return np.nan
    else:
        return value
#calling a function through applymap on the whole billboard dataframe    
bb = bb.applymap(replace_nulls)
```

How many NaNs are there in each of the week columns? If there are columns where they are all NaNs we can drop the columns using .dropna(how='all')

### The dataset is partially cleaned. Let's take a subset where we remove all the week positions columns. Then create a new dataframe of just weeks, where the method 'melt' will be used to calculate some key metrics

```python
clean_bb = bb.iloc[:,[0,1,2,4,6,7,8,9,10]]
clean_bb.head(2)
```

```python
melt = pd.melt(bb,id_vars=['artist','track'],value_vars=week_criteria, var_name = 'week_num',value_name = 'Position')
melt.dropna(inplace = True)
melt.sort_values(['artist','track'],inplace = True)
melt.reset_index(drop = True,inplace = True)

print melt.shape
#changing the data type of Position column from object to integer
melt['Position'] = melt['Position'].astype(int)
#Using apply function to change the week_num column from object to integer by string manipulation
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('_week','')).apply(lambda x: x.replace('x',''))
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('st','')).apply(lambda x: x.replace('nd',''))
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('rd','')).apply(lambda x: x.replace('th',''))
melt['week_num'] = melt['week_num'].astype(int)
print melt.dtypes
```

```python
melt.head()
```

```python
#creating a position bin, alotting top100 into a bins of 10.
pos = []
for x in melt['Position']:
    p = melt['Position']
    if (x >=1) and x <=10:
        pos.append(1)
    elif x>10 and x<=20:
        pos.append(2)
    elif x>20 and x<=30:
        pos.append(3)    
    elif x>30 and x<=40:
        pos.append(4)
    elif x>40 and x<=50:
        pos.append(5)
    elif x>50 and x<=60:
        pos.append(6)
    elif x>60 and x<=70:
        pos.append(7)
    elif x>70 and x<=80:
        pos.append(8)
    elif x>80 and x<=90:
        pos.append(9)
    else:
        pos.append(10)
        
melt['pos_cat'] = pd.Series(pos)
melt['pos_cat'].nunique()
```

```python
#Finding the Max, Min, and count of Position
pos_groupby = melt.groupby(['artist','track'])['Position'].agg([max,min,len])
pos_groupby.reset_index(inplace = True)
pos_groupby = pos_groupby.rename(columns = {'max':'lowest_pos','min': 'highest_pos','len':'weeks_in_chart'})
```

```python
melt2 = pd.merge(melt,pos_groupby)
melt2.head()
```

```python
#lets find the position entered, thus looking only at week 1
pos_entered = melt[melt['week_num'] == 1].groupby(['artist','track'])['Position'].max().to_frame('pos_entered')
pos_entered.reset_index(inplace = True)
melt3 = pd.merge(melt2,pos_entered)
melt3
```

```python
#finding the max week number of each artist track to find the position 
#when the track left the charts
pos_left = melt.groupby(['artist','track'])['week_num'].max().to_frame()
pos_left.reset_index(inplace = True)
pos_left
pos_left = pd.merge(pos_left,melt3.iloc[:,:4]).rename(columns = {'Position':'pos_left'})

melt4 = pd.merge(melt3,pos_left,how='left',on=['artist','track'])
del melt4['week_num_y']
melt4 = melt4.rename(columns = {'week_num_x':'week_num'})
```

```python
#lets try to find the week number at which the track peaked
melt5 = melt4[melt4['Position'] == melt4['highest_pos']].groupby(['artist','track'])['week_num'].min().to_frame('week_peaked')
melt5.reset_index(inplace = True)
melt5
melt6 = pd.merge(melt4,melt5)
#creating new variables to see increase in total positions, and decrease in total positions
melt6['climbed'] = melt6['pos_entered'] - melt6['highest_pos']
melt6['dropped'] = abs(melt6['highest_pos'] - melt6['pos_left'])
melt6.head(3)
```

```python
#these are the artist-tracks where the week number is greater than the count of weeks, meaning 
#these tracks dropped off and then came back on the chart
week_check = melt3.groupby(['artist','track'])['week_num','weeks_in_chart','highest_pos'].max()
week_check.reset_index(inplace = True)
week_check.head()
week_check['check'] = week_check['week_num'] - week_check['weeks_in_chart']
week_check = week_check[week_check['check'] > 0]
week_check.reset_index(drop = True,inplace = True)
print 'The Percentage of tracks that dropped off and then came back on is: {}%'.\
format(round((week_check.index[-1]+1)/float(len(clean_bb)),4)*100)

week_check
```

```python
#final dataset is the culmination of all previous melts used.
final = pd.merge(clean_bb,melt6)
final.head(2)
```

```python
#creating a subset to merge with the data frame with unique track & artist
#dropping all duplicate entries
melt_merge = melt6.iloc[:,[0,1,5,6,7,8,9,10,11,12]]
melt_merge = melt_merge[melt_merge.duplicated() == False]
melt_merge.reset_index(inplace = True, drop=True)
```

### Hypothesis - 

H<sub>0</sub>: Average Position of Pop Genre during the Summer is equal to the mean of all genres during the summer

H<sub>a</sub>: Pop Genre during the Summer is not equal to the mean of all genres during the summer

```python

summer = sub_final[sub_final['Season']=='Summer']['Position']
summer_notpop = sub_final[(sub_final['Season']=='Summer') & (sub_final['genre']!='Pop')]['Position']
summer_pop = sub_final[(sub_final['Season']=='Summer') & (sub_final['genre']=='Pop')]['Position']
print "The Average Position of All Tracks in the Summer is: ", summer.mean().round(2)
print "The Average Position of All minus Pop Genre Tracks in the Summer is: ", summer_notpop.mean().round(2)
print "The Average Position of Pop Genre Tracks in the Summer is: ", summer_pop.mean().round(2)
```

```python
#The p-value is 0.0001, which is a 100x lower than a value of alpha = 0.01
print stats.ttest_1samp(summer_pop,summer.mean())
```

From the T-Test above, testing the Null Hypothesis that the Average Position of the Pop tracks in the summer is equal to the Average Position of all tracks in the summer is rejected as it does not meet the threshold of alpha = 0.01
