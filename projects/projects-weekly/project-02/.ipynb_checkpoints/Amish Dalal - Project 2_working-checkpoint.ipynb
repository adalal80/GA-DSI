## Normalizing Track Length, Count Weeks columns, and Days_to_Peak column

Below we are trying to find if there is a difference between pandas and numpy calculation of mean and standard deviation for the track length column, respectively. The means are equal, but the standard deviation are off by 0.07. This seems negligible

```python
print bb['Track_length(s)'].mean(), np.mean(bb['Track_length(s)'])
print bb['Track_length(s)'].std(), np.std(bb['Track_length(s)'])

print 'The Pandas calculation of standard deviation is {}% higher than numpy\'s calculation of standard deviation'\
    .format(100 * (bb['Track_length(s)'].std() - np.std(bb['Track_length(s)']))/np.std(bb['Track_length(s)']))
```

Below we are trying to find if there is a difference between pandas and numpy calculation of mean and standard deviation for the track length column, respectively. The means are equal, but the standard deviation are off by 0.014. This seems negligible

```python
print bb['chart_weeks'].mean(), np.mean(bb['chart_weeks'])
print bb['chart_weeks'].std(), np.std(bb['chart_weeks'])

print 'The Pandas calculation of standard deviation is {}% higher than numpy\'s calculation of standard deviation'\
    .format(100 * (bb['chart_weeks'].std() - np.std(bb['chart_weeks']))/np.std(bb['chart_weeks']))
```

Given that the mean and standard deviations of pandas and numpy are quite similar, let's use numpy mean and standard deviation for normalizing the track length and chart weeks columns with z-score

```python
bb.iloc[:,:20].head(1)
```

```python
norm_tracklength = []
for x in bb['Track_length(s)']:
    z_track = (x - np.mean(bb['Track_length(s)']))/np.std(bb['Track_length(s)'])
    norm_tracklength.append(z_track)

insert_ztrack = bb.columns.get_loc('highest_rank') + 1
```

```python
bb.insert(insert_ztrack,'norm_tracklength',norm_tracklength)
```

```python
norm_chartweeks = []
for x in bb['chart_weeks']:
    z_weeks = (x - np.mean(bb['chart_weeks']))/np.std(bb['chart_weeks'])
    norm_chartweeks.append(z_weeks)

insert_zweeks = bb.columns.get_loc('norm_tracklength') + 1
```

```python
bb.insert(insert_zweeks,'norm_chartweeks',norm_chartweeks)
```

```python
norm_daystopeak = []
for x in bb['days_to_peak']:
    z_days = (x - np.mean(bb['days_to_peak']))/np.std(bb['days_to_peak'])
    norm_daystopeak.append(z_days)
norm_daystopeak
insert_zdays = bb.columns.get_loc('norm_chartweeks') + 1
```

```python
bb.insert(insert_zdays,'norm_days_to_peak',norm_daystopeak)
```

```python
melt = pd.melt(bb,id_vars=['artist','track'],value_vars=week_criteria)
melt.dropna(inplace = True)
melt.sort_values(['artist','track'],inplace = True)
melt.reset_index(drop = True,inplace = True)
```

```python
melt.rename(columns = {'value':'Position','variable':'week_num'},inplace = True)
```

```python
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('_week','')).apply(lambda x: x.replace('x',''))
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('st','')).apply(lambda x: x.replace('nd',''))
melt['week_num'] = melt['week_num'].apply(lambda x: x.replace('rd','')).apply(lambda x: x.replace('th',''))
melt['week_num'] = melt['week_num'].astype(int)
```

```python
melt.dtypes
melt.shape
```

```python
melt['Position'] = melt['Position'].astype(int)
```

```python
pos = []
for x in melt['Position']:
    p = melt['Position']
    if p >=1 and p <=10:
        pos.append(1)
    elif p>10 and p<=20:
        pos.append(2)
    elif p>20 and p<=30:
        pos.append(3)    
    elif p>30 and p<=40:
        pos.append(4)
    elif p>40 and p<=50:
        pos.append(5)
    elif p>50 and p<=60:
        pos.append(6)
    elif p>60 and p<=70:
        pos.append(7)
    elif p>70 and p<=80:
        pos.append(8)
    elif p>80 and p<=90:
        pos.append(9)
    else:
        pos.append(10)
print pos
```
