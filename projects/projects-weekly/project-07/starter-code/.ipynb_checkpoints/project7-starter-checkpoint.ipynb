Implementing PCA to investigate Airport Delays

### Problem Statement

This analysis investigates operations of major airports aroudn the country to understand the characteristics of departure and operational delays. 

- A certain degree of delay is expected in airport operations, however the FAA is noticing significant delays with certain airports
- When a flight takes off, it's departure delay is recorded in minutes, as well as operational data relating to this delay
- At the end of the year, this data is averaged out for each airport. Your datasets have these averaged for a 10 year range between 2004 and 2014
- Over this 10 year range, some delay times have not improved or have worsened.

#### Hypothesis:
- Taxi Out time (time it takes to leave the airport from gate) is associate with congestion time. The number of flights in and out of the airport would cause the taxi out time to increase as the number of runways is limited. This will play a huge factor in departure delays

#### Operations Dataset

```python
ops = pd.read_csv('../assets/Airport_operations.csv')
#renaming some columns for merge later on
ops = ops.rename(columns = {'year':'Year','airport':'Airport','departures for metric computation':'departures',\
                           'arrivals for metric computation':'arrivals'})
#putting year into datetime object
ops['Year'] = pd.to_datetime(ops['Year'], format='%Y').dt.year
print 'Operations dataframe shape is:',ops.shape
#analyzing to see how many airports have less than 11 years of data
avg_dep = ops.groupby('Airport')['departures','average airport departure delay'].mean().round()
avg_dep.reset_index(inplace = True)
avg_dep = avg_dep.rename(columns = {'average airport departure delay':'dep_delay'})
airport_count = ops.groupby(['Airport'])['Year'].count().to_frame().sort_values('Year',ascending = True)
#airports where there is less than  a count of 11
print airport_count[airport_count['Year'] < 11].index
print 'PSP and SMF are airports that have less than 11 years'
print 'median dep delay:',avg_dep['dep_delay'].median()
print 'mean dep delay:',avg_dep['dep_delay'].mean().round(0)

#major hubs, secondary hubs, spoke airports
major_hubs = ['ATL','ORD','DFW','DEN','LAX','IAH','CLT','DTW','PHL','PHX','MSP','EWR']
secondary_hubs = ['JFK','LGA','SFO','BOS','IAD','MIA','SLC','CVG','DCA','SEA']
hubs_1 = []
hubs_2 = []
spoke = []
ap_type = []
for x in avg_dep['Airport']:
    if x in major_hubs:
        ap_type.append('major')
        hubs_1.append(1)
        hubs_2.append(0)
        spoke.append(0)
    elif x in secondary_hubs:
        ap_type.append('secondary')
        hubs_1.append(0)
        hubs_2.append(1)
        spoke.append(0)
    else:
        ap_type.append('spoke')
        hubs_1.append(0)
        hubs_2.append(0)
        spoke.append(1)
# Can use major hubs, secondary hubs, and spoke as class labels
avg_dep['major_hubs'] = avg_dep['Airport'].apply(lambda x: 1 if x in major_hubs else 0)
avg_dep['secondary_hubs'] = avg_dep['Airport'].apply(lambda x: 1 if x in secondary_hubs else 0)
avg_dep['spoke'] = avg_dep['Airport'].apply(lambda x: 1 if (x not in secondary_hubs) & (x not in major_hubs) else 0)
avg_dep.sort_values('departures',ascending = False).head()
avg_dep['ap_cat'] = ap_type
#creating class for airport category
i=0
ap_cat = {}
for x in avg_dep['ap_cat'].unique():
    if x=='major':
        ap_cat[x]=0
    elif x=='secondary':
        ap_cat[x]=1
    else:
        ap_cat[x]=2
avg_dep['ap_class'] = avg_dep['ap_cat'].apply(lambda x: ap_cat[x])
avg_dep.head()

#get the airport - ap class
airport_class = avg_dep.iloc[:,[0,-1]]
```

```python
from matplotlib import cm
pca = PCA(n_components = 2)
X_trans = pca.fit_transform(X_scaledpca)
k = 8
kmeans = KMeans(n_clusters=k)
y_km = km.fit_predict(X_trans)
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X_trans,y_km)
yticks = []
y_ax_lower, y_ax_upper = 0,0
plt.subplots(figsize=(15,7))
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(i / n_clusters)

    plt.barh(range(y_ax_lower, y_ax_upper),c_silhouette_vals, height=1.0, edgecolor='none',color=color)
    yticks.append((y_ax_lower + y_ax_upper)/2)
    y_ax_lower += len(c_silhouette_vals)
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg,color='red',linestyle="--")
plt.yticks(yticks,cluster_labels + 1)
plt.ylabel("Cluster",fontsize = 16)
plt.xlabel("Silhouette Coefficient",fontsize = 16)
plt.title("Silhouette Scores",fontsize = 20)
plt.show()
```
