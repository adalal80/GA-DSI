##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

First, lets take out the 'All' observations, which is done below. Below that are functions for the mean and standard deviation.

```python
print state
new_state = state[1:len(state)-1]
print new_state

new_iVerbal = iVerbal[1:len(iVerbal)-1]
print new_iVerbal
len(new_iVerbal)

new_iRate = iRate[1:len(iRate)-1]
print new_iRate
len(new_iRate)

new_iMath = iMath[1:len(iMath)-1]
print new_iMath
len(new_iMath)
```

```python
def mean_d(col):
    n = len(col)
    return sum(col)/float(n)

def std(col):
    n = len(col)
    c = sum(col)/float(n)
    ss = np.sum((x-c)**2 for x in col)
    std = (ss/n)**.5
    return std
```

```python
np.std(new_iRate), std(new_iRate), np.mean(new_iRate), mean_d(new_iRate)
```

```python
np.std(new_iVerbal), std(new_iVerbal), np.mean(new_iVerbal), mean_d(new_iVerbal)
```

```python
np.std(new_iMath), std(new_iMath), np.mean(new_iMath), mean_d(new_iMath)
```

As you can see, the calculated mean and standard deviations from the function are very close approximations to the numpy mean and standard deviation.
