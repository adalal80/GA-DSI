### ZIllow

```python
URL = "http://www.zillow.com/brooklyn-new-york-ny-11249/rentals/"
z = requests.get(URL)
soup_z = BeautifulSoup(z.text,"lxml")
```

```python
print soup_z.prettify()
```

```python
for x in  soup_z.findAll('div',{"class":"minibubble template hide"}):
    a = str(x).split(",")
    for x in a:
        if "bed" in x:
            x = x.replace("<div class=\"minibubble template hide\"><!--{","")
            print x
#         if "bed" in x:
#             print x.split("{")
```

```python
#sqft
for x in  soup_z.findAll('div',{"class":"minibubble template hide"}):
    print str(x).split("sqft")[-1].split(',')[0]
```

```python
#bedroom
for x in  soup_z.findAll('span',{"class":"zsg-photo-card-unit"}):
    print x.text.split('$')[0]
    
#price
for x in  soup_z.findAll('span',{"class":"zsg-photo-card-unit"}):
    print x.text.split('$')[1].replace('+','')
    
#link
for x in  soup_z.findAll('a',{"class":"zsg-photo-card-overlay-link routable hdp-link"}):
    print x["href"]
```
